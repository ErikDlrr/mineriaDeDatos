IPython Magic Commands

```python
# El comado mágico %timeit sirve para evaluar el
# desempeño, en términos de tiempo
%timeit L = [n ** 2 for n in range(1000)]
```

```python
# El comando mágico %%timeit sirve también para evaluar el desempeño
# solo que este abarca varias líneas de código en lugar de una
%%timeit
L = []
for n in range(1000):
    L.append(n ** 2)
```
